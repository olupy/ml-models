```python
import pandas as pd
import io
import matplotlib.pyplot as plt
from google.colab import files

# Load the data from Excel file
uploaded = files.upload()
data = pd.read_excel(io.BytesIO(uploaded['JanData.xlsx']))

# Split the data based on locations
majidun_data = data[data['loc'] == 'Majidun']
motorwatay_data = data[data['loc'] == 'Motorway']
mile12_data = data[data['loc'] == 'Mile-12']

# Extract a 24-hour period of data
time_period = pd.date_range(start='/01/01/2019 0:00', periods=24, freq='H')
majidun_24h = majidun_data[majidun_data['date'].isin(time_period)]
motorwatay_24h = motorwatay_data[motorwatay_data['date'].isin(time_period)]
mile12_24h = mile12_data[mile12_data['date'].isin(time_period)]

# Plot PM2.5 and PM10 variation for Majidun
plt.figure(figsize=(12, 6))
plt.plot(majidun_24h['date'], majidun_24h['pm25'], label='PM2.5')
plt.plot(majidun_24h['date'], majidun_24h['pm10'], label='PM10')
plt.title('PM2.5 and PM10 Variation in Majidun')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Plot PM2.5 and PM10 variation for Motorwatay
plt.figure(figsize=(12, 6))
plt.plot(motorwatay_24h['date'], motorwatay_24h['pm25'], label='PM2.5')
plt.plot(motorwatay_24h['date'], motorwatay_24h['pm10'], label='PM10')
plt.title('PM2.5 and PM10 Variation in Motorwatay')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend()
plt.xticks(rotation=45)
plt.show()

# Plot PM2.5 and PM10 variation for Mile-12
plt.figure(figsize=(12, 6))
plt.plot(mile12_24h['date'], mile12_24h['pm25'], label='PM2.5')
plt.plot(mile12_24h['date'], mile12_24h['pm10'], label='PM10')
plt.title('PM2.5 and PM10 Variation in Mile-12')
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
